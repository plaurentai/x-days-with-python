# Combining Data Structure

### One Multidimensional List

```python
# Daily high abnd low temperatures for a week (in fahrenheit)
temperature = [
    [20, 25],
    [15, 18],
    [10, 12]
]
```

```python
# Day 1 temperature
print(temperature[0])  # Output: [20, 25]

# Day 2 low temperature
print(temperature[1])  # Output: [15, 18]

# Day 3 temperature
print(temperature[2])  # Output: [10, 12]

# Day 1 high temperature
print(temperature[0][1])  # Output: 25

# Day 2 low temperature
print(temperature[1][0])  # Output: 15

# Day 2 high temperature
print(temperature[1][1])  # Output: 18
```

### Two dimensional list    

```python
temperatures = [
    [
        [20, 25],
        [15, 18],
        [10, 12]
    ],
    [
        [30, 35],
        [25, 28],
        [20, 22]
    ]
]
```

```python
print(temperatures[0][0])  # Output: [20, 25]
print(temperatures[0][1][1])  # Output: 18
```
