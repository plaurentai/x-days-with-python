# IF

```python
a = "Yes"
b = "No"

if a == b:
    print("a is equal to b")
else:
    print("a is not equal to b")
```

```python
a = 1
b = 2

if a < b:
    print("a is less than b")
else:
    print("a is not less than b")
```

## Elseif

```python
a = 1
b = 2
c = 3

if a > b:
    print("a is greater than b")
elif b < c:
    print("b is less than c")
```
